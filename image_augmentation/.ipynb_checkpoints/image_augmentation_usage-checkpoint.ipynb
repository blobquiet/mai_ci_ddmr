```python
import matplotlib.pyplot as plt
```

```python
# Load images from folder in format
import glob
folder_name= "test/benign"
img_format= "jpg"
images=glob.glob("./" + folder_name + "/*."+ img_format)
```

```python
from skimage.io import imread
from random import randrange
# Usage
# from ImageAugmentation import Augmentation
```

```python
from ImageAugmentation import Augmentation
img=imread(images[randrange(10)])/255

aug = Augmentation(img)

fig, axs = plt.subplots(1, 2, figsize=(10,10))
axs[0].axis('off') #removing axis
axs[0].imshow(img);
axs[0].set_title('original image')
axs[1].axis('off') #removing axis
axs[1].imshow(aug);
axs[1].set_title('augmented image')
plt.show()
```

```python
# Usage for multiple images in a list for example
Aug_im=[]
Org_im=[]
# number of augmented changes to image
changes = 5
count=0
for img_path in images:
    image = imread(img_path)
    original_im= Org_im.append(image)
    Aug_im.append([Augmentation(image) for i in range(changes)])
    count=count+1
    if count==5:
        break
```

```python
selected_image=1
new = Augmentation(Aug_im[selected_image][1])

plt.imshow(Org_im[selected_image])
plt.title('original image')
plt.axis('off')
plt.show()

fig, axs = plt.subplots(1, changes, figsize=(15,15))
for i in range(changes):
    axs[i].axis('off') #removing axis
    axs[i].imshow(Aug_im[selected_image][i], cmap='gray');
    axs[i].set_title('change '+str(i))
plt.show()
```
