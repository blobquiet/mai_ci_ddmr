```python
import matplotlib.pyplot as plt
```

```python
# Load images from folder in format
import glob
folder_name= "test/benign"
img_format= "jpg"
images=glob.glob("./" + folder_name + "/*."+ img_format)
```

```python
from skimage.io import imread
# Usage
from ImageAugmentation import Augmentation
img=imread(images[3])/255
```

```python
new = Augmentation(img)
plt.imshow(new)
plt.axis('off')
plt.show()
```

```python
# Usage for multiple images in a list for example
Aug_im=[]
# number of augmented changes to image
changes = 10
count=0
for img_path in images:
    image = imread(img_path)
    Aug_im.append([Augmentation(image) for i in range(changes)])
    count=count+1
    if count==5:
        break
```

```python
new = Augmentation(Aug_im[3][5])
plt.imshow(new)
plt.axis('off')
plt.show()
```
